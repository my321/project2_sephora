```python
features_details=[]
features=[]

if soup.find('div',{'class': 'css-pz80c5'}).find_all('b')[0].text == "Which skin type is it good for?":
    for div in soup.find('div',{'class': 'css-pz80c5'}).find_all('b'):
        features.extend([div.text.strip()])
        features_details.extend([div.next_sibling.next_sibling])
else:
    for div in soup.find('div',{'class': 'css-pz80c5'}).find_all('b'):
        features.extend([div.text.strip()])
        features_details.extend([div.next_sibling.strip()])

final_features=[]
for item in features: 
    final_features.extend([item.replace(':','')])
    

final_features
```

```python
if "Clean at Sephora" in final_features:
    Clean_Sephora="yes"
else:
    Clean_Sephora="no"
```

```python
Clean_Sephora
```

```python
import pandas as pd
df=pd.DataFrame([features_details], columns=final_features)
df
```

```python
skin_types=[]
skin=[]
if 'Skin Type' in final_features:
    for item in df['Skin Type'][0].split(","):
        skin_types.extend([item.replace('and','').strip()])  
else: 
    eldest=soup.find('div',{'class': 'css-pz80c5'}).find_all('b')[0]
    younger=''
    name=''
    while name!='b':
        younger=eldest.next_sibling
        name=younger.name
        if name==None:
            skin.append(younger)
        eldest=younger
    if "\n" in skin[-1]:        
        for i in range(0,len(skin)-1):
            skin_types.append(skin[i].split(" ")[1])
        skin_types.append(skin[-1].split(" ")[2].replace('\n',''))
    else: 
        for i in range(0,len(skin)):
            skin_types.append(skin[i].split(" ")[1])

    
skin_types
```
