## Using the list of individual product urls and the code from all the practice scraps, scraping each products page
<p> First created a function, get_product_dict, to run on each individual url that scraps the data and creates a dictionary for the product. Then ran a for loop to call each of the product urls and run the function get_product_dict on each url. Then append a master list with the individual product dictionaries. Finally, save the master list  using JSON. </p>

```python
driver = webdriver.Chrome(chromedriver)
def get_product_dict(link):
    '''
    From Sephora product link stub, request product html, parse with BeautifulSoup, and
    collect 
        - product 
        - brand
        - size 
        - price
        - formulation
        - for what skin type
        - other product details     
    Return information as a dictionary.
    '''
    base_url ='https://www.sephora.com'
    url = base_url + link
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    soup.prettify()
    headers = ['product', 'brand','review', 'loves', 'number of reviews',
           'price', 'size ounces', 'size mL', 'formulation', 'clean at sephora', 
           'oily', 'dry', 'combination', 'sensitive', 'normal', 'vegan',
          'cruelty-free', 'gluten-free', 'recyclable packaging']
 
    #get product and brand 
    product = soup.find('title').text.split(" | ")[0].split(" - ")[0].strip()
    brand = soup.find('title').text.split(" | ")[0].split(" - ")[1].strip()
    
    #get review
    review=soup.find('div', {'class':"css-jp4jy6"})['aria-label'].replace('stars','').strip()
    
    #get loves and number of reviews 
    if (soup.find_all(class_="css-2rg6q7")) is False: 
        if 'K' in soup.find(class_="css-1coslxg").text: 
            num_reviews=float(soup.find(class_="css-1coslxg").text.replace('K',''))*1000
        else: 
            num_reviews=float(soup.find(class_="css-1coslxg").text)
        
        if 'K' in soup.find(class_="css-jk94q9").text:
            loves=float(soup.find(class_="css-jk94q9").text.replace('K',""))*1000
        else: 
            loves=float(soup.find(class_="css-jk94q9").text)
        
    if bool(soup.find_all(class_="css-2rg6q7")):
        if 'K' in soup.find_all(class_="css-2rg6q7")[1].text.split(" ")[0]:
            loves=float(soup.find_all(class_="css-2rg6q7")[1].text.split(" ")[0].replace("K",''))*1000
        else: 
            loves=float(soup.find_all(class_="css-2rg6q7")[1].text.split(" ")[0])
        if 'K' in soup.find_all(class_='css-2rg6q7')[0].text.split(" ")[0]:
            num_reviews=float(soup.find_all(class_='css-2rg6q7')[0].text.split(" ")[0].replace('K',''))*1000
        else: 
            num_reviews=float(soup.find_all(class_='css-2rg6q7')[0].text.split(" ")[0])
   
    #get price 
    price=float(soup.find_all('span',{'data-at':'price'})[0].text.replace('$','').strip())
    
    #get size ounces and mL
    if (bool(soup.find_all('span',{'data-at':'item_variation_type'}))) and ("SIZE" in (soup.find_all('span',{'data-at':'item_variation_type'})[0].text)):
        if "fl. oz" in (soup.find_all('span',{'data-at':'item_variation_type'})[0].text):
            size_oz=float(soup.find_all('span',{'data-at':'item_variation_type'})[0].text.replace('SIZE: ','').split('/')[0].replace("fl. oz",'').strip())
            size_ml=float(soup.find_all('span',{'data-at':'item_variation_type'})[0].text.replace('SIZE: ','').split('/')[1].replace("mL",'').strip())
       
        else:
            size_oz=float(soup.find_all('span',{'data-at':'item_variation_type'})[0].text.replace('SIZE: ','').split('/')[0].replace("oz",'').strip())
            size_ml=float(soup.find_all('span',{'data-at':'item_variation_type'})[0].text.replace('SIZE: ','').split('/')[1].replace("mL",'').strip())
     
    
    if (bool(soup.find_all('div',{'data-at':"sku_size"}))) and ("SIZE" in (soup.find_all('div',{'data-at':"sku_size"})[0].text)):
        size_oz=float(soup.find_all('div',{'data-at':"sku_size"})[0].text.replace('SIZE ','').split("ITEM")[0].split('/')[0].split(' ')[0].strip())
        size_ml=float(soup.find_all('div',{'data-at':"sku_size"})[0].text.replace('SIZE ','').split("ITEM")[0].split('/')[1].strip().split(' ')[0].strip())

    if ((bool(soup.find_all('span',{'data-at':'item_variation_type'})) and ("SIZE" in soup.find_all('span',{'data-at':'item_variation_type'})[0].text)) is False) and ((bool(soup.find_all('div',{'data-at':"sku_size"})) and ("SIZE" in soup.find_all('div',{'data-at':"sku_size"})[0].text)) is False):
        size_oz=None  
        size_ml=None

        
    #get table for remaining features 
    features_details=[]
    features=[]


    if soup.find('div',{'class': 'css-pz80c5'}).find_all('b')[0].text == "Which skin type is it good for?":
        for div in soup.find('div',{'class': 'css-pz80c5'}).find_all('b'):
            features.extend([div.text.strip()])
            features_details.extend([div.next_sibling.next_sibling])
    else:
        for div in soup.find('div',{'class': 'css-pz80c5'}).find_all('b'):
            features.extend([div.text.strip()])
            features_details.extend([div.next_sibling])
        
    final_features=[]
    for item in features: 
        final_features.extend([item.replace(':','')])
    df=pd.DataFrame([features_details], columns=final_features)
    
    #get clean at sephora 
    if "Clean at Sephora" in final_features:
        Clean_Sephora="yes"
    else:
        Clean_Sephora="no"
    
    #get skin type 
    skin_types=[]
    skin=[]
    i=0
    if 'Skin Type' in final_features:
        for item in df['Skin Type'][0].split(","):
            skin_types.extend([item.replace('and','').strip()])
    if 'Which skin type is it good for?' in final_features: 
        eldest=soup.find('div',{'class': 'css-pz80c5'}).find('b')
        younger=''
        name=''
        while name!='b':
            younger=eldest.next_sibling
            name=younger.name
            if name==None:
                skin.append(younger)
            eldest=younger
            i+=1
            if i>10:
                break
        if "\n" in skin[-1]:        
            for i in range(0,len(skin)-1):
                skin_types.append(skin[i].split(" ")[1])
            skin_types.append(skin[-1].split(" ")[2].replace('\n',''))
        else:    
            for i in range(0,len(skin)):
                skin_types.append(skin[i].split(" ")[1])
    #get variables for skin type
    if 'Combination' in skin_types:
        for_combination='yes'
    else:
        for_combination='no'
        
    if 'Normal' in skin_types:
        for_normal='yes'
    else: 
        for_normal='no'
    
    if 'Oily' in skin_types:
        for_oily='yes'
    else: 
        for_oily='no'
    
    if 'Sensitve' in skin_types:
        for_sensitive='yes'
    else:
        for_sensitive='no'
    
    if 'Dry' in skin_types:
        for_dry="yes"
    else: 
        for_dry='no'
    
    #get formulation
    
    if "Formulation" in final_features: 
        formulation=df['Formulation'][0].strip()
    else: 
        formulation='moisturizer'
    
    #get other product details 
    
    if 'Ingredient Callouts' in final_features: 
    
        if 'vegan' in df['Ingredient Callouts'][0]:
            vegan='yes'
        else: 
            vegan='no'
    
        if 'cruelty-free' in df['Ingredient Callouts'][0]:
            cruelty_free='yes'
        else: 
            cruelty_free='no'
    
        if 'gluten-free' in df['Ingredient Callouts'][0]:
            gluten_free='yes'
        else: 
            gluten_free='no'
    
        if 'recyclable packaging' in df['Ingredient Callouts'][0]:
            recycle='yes'
        else: 
            recycle='no'
    
    else: 
        vegan, cruelty_free, gluten_free, recycle = 'no','no','no','no'
  
    product_dict = dict(zip(headers, [product,
                                brand,
                                review,
                                loves, 
                                num_reviews,
                                 price,
                                 size_oz,
                                 size_ml,
                                 formulation,
                                 Clean_Sephora,
                                 for_oily,
                                 for_dry, 
                                 for_combination,
                                 for_sensitive,
                                 for_normal, 
                                 vegan, 
                                 cruelty_free, 
                                 gluten_free, 
                                 recycle]))
    return product_dict
        
     

    
```
